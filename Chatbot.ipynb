```python
import nltk
import random
import string
import re
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from IPython.display import display
import ipywidgets as widgets

# Download NLTK data
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')

print("âœ… Libraries loaded successfully!")
```

```python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
#  Knowledge Base: intents with patterns & responses
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

intents = {
    "greeting": {
        "patterns": ["hi", "hello", "hey", "good morning", "good evening", "howdy", "sup"],
        "responses": [
            "Hey there! How can I help you? ğŸ˜Š",
            "Hello! What can I do for you today?",
            "Hi! Feel free to ask me anything."
        ]
    },
    "goodbye": {
        "patterns": ["bye", "goodbye", "see you", "take care", "later", "exit", "quit"],
        "responses": [
            "Goodbye! Have a great day! ğŸ‘‹",
            "See you later! Take care!",
            "Bye! Come back anytime."
        ]
    },
    "thanks": {
        "patterns": ["thanks", "thank you", "thx", "appreciate", "thank u"],
        "responses": [
            "You're welcome! ğŸ˜Š",
            "Happy to help!",
            "No problem at all!"
        ]
    },
    "name": {
        "patterns": ["what is your name", "who are you", "your name", "what should i call you"],
        "responses": [
            "I'm SimpleBot, your friendly AI assistant! ğŸ¤–",
            "My name is SimpleBot. Nice to meet you!",
        ]
    },
    "how_are_you": {
        "patterns": ["how are you", "how do you do", "are you okay", "how are you doing"],
        "responses": [
            "I'm just a bot, but I'm doing great! Thanks for asking ğŸ˜„",
            "Feeling chatty and ready to help!",
        ]
    },
    "capabilities": {
        "patterns": ["what can you do", "help me", "what do you know", "your features", "capabilities"],
        "responses": [
            "I can answer questions, have small conversations, tell jokes, and share fun facts! Try asking me something ğŸ™‚",
            "I can chat, answer FAQs, tell jokes, and share facts. What would you like?"
        ]
    },
    "joke": {
        "patterns": ["tell me a joke", "say something funny", "joke", "make me laugh", "funny"],
        "responses": [
            "Why do programmers prefer dark mode? Because light attracts bugs! ğŸ›ğŸ˜‚",
            "Why did the computer go to the doctor? Because it had a virus! ğŸ’»ğŸ˜·",
            "I told my computer I needed a break. Now it won't stop sending me Kit-Kat ads! ğŸ˜„"
        ]
    },
    "fact": {
        "patterns": ["tell me a fact", "interesting fact", "did you know", "fun fact", "share a fact"],
        "responses": [
            "Did you know? Honey never spoils. Archaeologists found 3000-year-old honey in Egyptian tombs! ğŸ¯",
            "Fun fact: Octopuses have three hearts and blue blood! ğŸ™",
            "Did you know? A bolt of lightning is five times hotter than the surface of the sun! âš¡"
        ]
    },
    "age": {
        "patterns": ["how old are you", "your age", "when were you born", "when were you created"],
        "responses": [
            "I was just created â€” I'm brand new! Age is just a number anyway ğŸ˜„",
            "I don't have a birthday, but I was built recently!"
        ]
    },
    "weather": {
        "patterns": ["what is the weather", "weather today", "is it raining", "temperature outside"],
        "responses": [
            "I can't check live weather yet, but you can visit weather.com for updates! ğŸŒ¤ï¸",
            "I don't have internet access for real-time weather, try Google for the latest forecast!"
        ]
    },
    "time": {
        "patterns": ["what time is it", "current time", "what is the time", "tell me the time"],
        "responses": ["__time__"]  # special tag handled in engine
    },
    "default": {
        "patterns": [],
        "responses": [
            "Hmm, I'm not sure about that. Could you rephrase? ğŸ¤”",
            "I didn't quite catch that. Try asking something else!",
            "That's beyond my knowledge for now. Ask me a joke or a fun fact instead! ğŸ˜Š"
        ]
    }
}

print("âœ… Knowledge base loaded with", len(intents) - 1, "intents!")
```

```python
from datetime import datetime

lemmatizer = WordNetLemmatizer()

def preprocess(text):
    """Lowercase, remove punctuation, tokenize, and lemmatize."""
    text = text.lower().translate(str.maketrans('', '', string.punctuation))
    tokens = word_tokenize(text)
    return [lemmatizer.lemmatize(token) for token in tokens]

def match_intent(user_input):
    """Match user input to the best intent using token overlap."""
    user_tokens = set(preprocess(user_input))
    best_intent = "default"
    best_score = 0

    for intent, data in intents.items():
        if intent == "default":
            continue
        for pattern in data["patterns"]:
            pattern_tokens = set(preprocess(pattern))
            # Jaccard-like overlap score
            overlap = len(user_tokens & pattern_tokens)
            if overlap > best_score:
                best_score = overlap
                best_intent = intent

    return best_intent

def get_response(intent_name):
    """Pick a random response from the matched intent."""
    responses = intents[intent_name]["responses"]
    response = random.choice(responses)

    # Handle special dynamic tags
    if response == "__time__":
        now = datetime.now().strftime("%I:%M %p")
        return f"The current time is {now} ğŸ•"

    return response

def chatbot_reply(user_input):
    """Main chatbot function: input â†’ response."""
    if not user_input.strip():
        return "Please type something! I'm here to chat ğŸ˜Š"
    intent = match_intent(user_input)
    return get_response(intent)

print("âœ… Chatbot engine ready!")
```

```python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
#  Quick unit tests â€” run a few test inputs
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

test_inputs = [
    "Hello!",
    "What is your name?",
    "Tell me a joke",
    "Did you know any fun facts?",
    "What can you do?",
    "What time is it?",
    "Goodbye!"
]

print("=" * 45)
print("  ğŸ¤– SimpleBot â€” Quick Test")
print("=" * 45)
for inp in test_inputs:
    print(f"You   : {inp}")
    print(f"Bot   : {chatbot_reply(inp)}")
    print("-" * 45)
```

```python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
#  Interactive Chat UI using ipywidgets
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

chat_history = widgets.Output(
    layout=widgets.Layout(
        border='1px solid #ccc',
        padding='10px',
        height='320px',
        overflow_y='scroll',
        background_color='#f9f9f9'
    )
)

input_box = widgets.Text(
    placeholder='Type your message here...',
    layout=widgets.Layout(width='75%')
)

send_btn = widgets.Button(
    description='Send ğŸ’¬',
    button_style='primary',
    layout=widgets.Layout(width='20%')
)

clear_btn = widgets.Button(
    description='Clear ğŸ—‘ï¸',
    button_style='warning',
    layout=widgets.Layout(width='10%')
)

def on_send(b):
    user_msg = input_box.value.strip()
    if not user_msg:
        return
    bot_reply = chatbot_reply(user_msg)
    with chat_history:
        print(f"ğŸ§‘ You : {user_msg}")
        print(f"ğŸ¤– Bot : {bot_reply}")
        print("-" * 50)
    input_box.value = ''

def on_clear(b):
    chat_history.clear_output()

send_btn.on_click(on_send)
clear_btn.on_click(on_clear)
input_box.on_submit(on_send)  # Press Enter to send

display(
    widgets.HTML("<h3>ğŸ¤– SimpleBot â€” Chat Interface</h3>"),
    chat_history,
    widgets.HBox([input_box, send_btn, clear_btn])
)
```

```python
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
#  Add your own custom FAQ intent at runtime
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

def add_custom_faq(intent_name, patterns, responses):
    """Dynamically add a new intent to the chatbot."""
    intents[intent_name] = {
        "patterns": patterns,
        "responses": responses
    }
    print(f"âœ… Intent '{intent_name}' added with {len(patterns)} pattern(s)!")

# Example: add a Python FAQ
add_custom_faq(
    intent_name="python_faq",
    patterns=["what is python", "tell me about python", "python programming"],
    responses=[
        "Python is a high-level, easy-to-read programming language popular in AI, web dev, and data science! ğŸ",
        "Python is loved for its simplicity and huge ecosystem of libraries like NumPy, Pandas, and TensorFlow!"
    ]
)

# Test it
print(chatbot_reply("tell me about python"))
```
